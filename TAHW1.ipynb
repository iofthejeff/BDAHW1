```python
import nltk
from nltk import FreqDist
import pandas
colnames = ['year', 'season', 'author', 'location', 'time', 'review_text']
raw = pandas.read_csv('/Users/jared/Downloads/fashion data.csv', names=colnames)
review = raw.review_text.tolist()
reviewstring = ' '.join([str(r) for r in review]) 
allclean = reviewstring.replace("'",' ')
allclean = allclean.replace('\n',' ')
wordsbag = allclean.strip().split()
words1 = [w.lower() for w in wordsbag]
words2 = [w for w in words1 if w.isalpha()]
tokens = nltk.word_tokenize(reviewstring)
```

```python
wordsfreq = FreqDist(words2)
sortedwords = sorted(wordsfreq.items(),key = lambda k:k[1], reverse = True)
sortedwords
```

```python
wordsfreq.plot(30)
```

```python
'''Method2'''
```

```python
from nltk.corpus import stopwords

stopwords = stopwords.words('english')
```

```python
nostopwords = [w for w in words2 if w not in stopwords]
```

```python
porter = nltk.PorterStemmer()
stemmednostopwords = [porter.stem(w) for w in nostopwords]
```

```python
stemmednostopwordsfreq = FreqDist(stemmednostopwords)
sortednostwords = sorted(stemmednostopwordsfreq.items(),key = lambda k:k[1], reverse = True)
sortednostwords
```

```python
stemmednostwordsfreq.plot(30)
```

```python
'''Method3'''
```

```python
POS_tags = nltk.pos_tag(tokens)
```

```python
POS_tags_N = [(word,tag) for (word,tag) in POS_tags if tag.startswith('NN')]
Nfreq = nltk.FreqDist(POS_tags_N)
sorted_Nfreq = sorted(Nfreq.items(), key = lambda k:k[1], reverse = True)
sorted_Nfreq
```

```python
Nfreq.plot(30)
```
